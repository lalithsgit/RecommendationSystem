```python
# Load packages
import pandas as pd
import numpy as np
from math import sqrt
import breakinghits_script as bh
import pickle
from scipy.sparse import csr_matrix
```

```python
# Implicit Recommendations for spotlights and saves
spotlights = bh.get_spotlights() # dataframe of user spotlighted songs
saves = bh.get_saves()           # dataframe of user saved songs
sparse_spots = bh.get_sparse(spotlights)
sparse_saves = bh.get_sparse(saves)
implicit, user_spot_save = bh.ensemble_implicit_recommendations(sparse_spots, sparse_saves,27,drop=False)
```

```python
rate = bh.pull_dataframe('select a.user_id, a.user_music_id, a.rating from user_music_votes a join user_musics b on a.user_music_id = b.id')
svd = bh.bh_model(rate)
predictions = bh.bh_predict(svd, 27, rate)
explicit = bh.bh_top_recs(predictions)
```

```python
recommendation_dict = {}
for i in user_list:
    print(i)
    implicit, user_spot_save = bh.ensemble_implicit_recommendations(sparse_spots, sparse_saves,i,drop=False)
    predictions = bh.bh_predict(svd, i, rate)
    explicit = bh.bh_top_recs(predictions)
    implicit['song_id'] = implicit.index.values
    implicit.columns = ['spotlight_saved', 'Song_ID']
    total = implicit.merge(explicit, how='left',on='Song_ID').replace(np.nan, 0)
    total['final_sum'] = total['spotlight_saved'] + total['rating']
    total.index = total['Song_ID']
    user_rated = np.unique(np.append(np.array(rate[rate['user_id'] == i]['user_music_id']),(user_spot_save)))
    recommendation_dict[i] = list(total.drop(user_rated).nlargest(10,['final_sum'])['Song_ID'])

pickle_out = open("recommendation.pkl","wb")
pickle.dump(recommendation_dict,pickle_out)
pickle_out.close

pickle_in = open("recommendation.pkl","rb")
example_dict = pickle.load(pickle_in)
```
